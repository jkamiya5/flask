```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg
import random
import string
import os
import pandas as pd
from matplotlib.backends.backend_agg import FigureCanvasAgg
import numpy as np
```

```python
player = "hamilton"
df = pd.read_csv('Kings_Game_HIST_' + player + '.csv', encoding='shift-jis')
df2 = pd.read_csv('Kings_Game_HISTORY.csv', encoding='shift-jis')
left = pd.DataFrame(df)
right = pd.DataFrame(df2)
output = pd.merge(left, right, how='left', on=['Time'])
```

```python
lose = output.where(output.WinLose == 0)
win = output.where(output.WinLose == 1)
win = win.dropna(subset=['No'])
lose = lose.dropna(subset=['No'])
```

```python
win.columns
```

```python
win = output.loc[:, ['WinLose', 'PTS', 'TRB', 'AST', 'ST', 'BLK', 'BSR', 'TOV', 'PF','FO']]
```

```python
win = win[0:35]
win
```

```python
ppp = pd.tools.plotting.scatter_matrix(win, diagonal="kde", figsize=(10,10))
plt.show()
ppp
```

```python
win
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA().fit(win)
pca
```

```python
print('load: ', pca.explained_variance_ratio_)
```

```python
print('cum load: ', pca.explained_variance_ratio_.cumsum())
```

```python
z = pca.transform(win)
```

```python
plt.scatter(z[:, 0], z[:, 1], s=100)
plt.show()
```

```python
fig, ax = plt.subplots()
```

```python
ax.bar(np.arange(10), pca.components_[0])
ax.set_xticks(np.arange(10) + 0.4)
ax.set_xticklabels(win.columns)
plt.show()
```

```python
from sklearn import cluster
kmeans = cluster.KMeans(4).fit(win)
print(kmeans.cluster_centers_)
```

```python
labels = kmeans.predict(win)
print(labels)
```

```python
cz = pca.transform(kmeans.cluster_centers_)
colors = ["b", "g", "r", "c"]
plt.scatter(z[:, 0], z[:, 1], s=100, c=[colors[i] for i in labels])
plt.scatter(cz[:, 0], cz[:, 1], s=1000, c="orange", marker="*")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from sklearn import model_selection
```

```python
win
```

```python
y = win["WinLose"]
y = y.replace('0.0', 'Lose')
y = y.replace(1, 'Win')
y
```

```python
X = win.iloc[:,1:15]
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(X.values, y.values, test_size=0.1)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(3).fit(x_train, y_train)
```

```python
for y_pred, y_true in zip(knn.predict(x_test), y_test):
    print(y_pred, y_true)
```

```python
print(knn.score(x_test, y_test))
```

```python
scores = model_selection.cross_val_score(KNeighborsClassifier(3), X, y, cv=5)
mean_score = scores.mean()
print(mean_score)
```

```python
for k in range(1, 11):
 score = model_selection.cross_val_score(KNeighborsClassifier(k), X, y, cv=5)
 mean_score = scores.mean()
 print(k, mean_score)
```
